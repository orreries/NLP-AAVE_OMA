#### This code was used to scrape the AAVE terms and accompanying definitions from the AAVE glossary website. The derived dataset is used for the AAVE dictionary aspect of my AAVE Semantic Similarity Project.

```python
import requests
from bs4 import BeautifulSoup
import lxml
import pandas as pd
```

```python
# save website data
site_terms = requests.get("https://aaveglossary.carrd.co/#a-e") # minstrelsy database (posters & handbills)
html_code = site_terms.content # saves the HTML code
soup_terms = BeautifulSoup(html_code) # create soup object
```

```python
# empty list for terms
aave_terms = []
term_def = []


# div id="list01"
div = soup_terms.find("div", {"id": "list01"})
div2 = soup_terms.find("div", {"id": "list02"})
div3 = soup_terms.find("div", {"id": "list03"})
div4 = soup_terms.find("div", {"id": "list04"})

# goes through to get nested strong (term) and mark (definition) elements
if div:
    for li in div.find_all("li"):
        for p in li.find_all("p"):
            for strong in p.find_all("strong"):
                aave_terms.append(strong.get_text(strip=True))
            for mark in p.find_all("mark"):
                term_def.append(mark.get_text(strip=True))
```

```python
aave_terms1 = []
term_def1 = []

# looping thru div that listXX class
for div in soup_terms.find_all("div", class_="list"):
    for li in div.find_all("li"):
        for p in li.find_all("p"):
            for strong in p.find_all("strong"):
                aave_terms1.append(strong.get_text(strip=True))
            for mark in p.find_all("mark"):
                term_def1.append(mark.get_text(strip=True))
```

```python
# make dataset where the word/def are properly attributed
term_def_pairs = []

for div in soup_terms.find_all("div", class_="list"):
    for li in div.find_all("li"):
        for p in li.find_all("p"):
            strong = p.find("strong")
            mark = p.find("mark")

            # only append if both exist
            if strong and mark:
                term_def_pairs.append({
                    "term": strong.get_text(strip=True),
                    "definition": mark.get_text(strip=True)
                })
```

```python
df = pd.DataFrame(term_def_pairs)
df
```

```python
# create csv
df.to_csv('AAVE-with-Definitions.csv')
```
